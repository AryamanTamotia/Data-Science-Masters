## Q1

The GET method requests data from a specified resource. It is a safe and idempotent method, which means that multiple identical requests should have the same effect as a single request. In other words, a GET request retrieves data without changing or modifying anything on the server. GET requests should only be used for safe operations, such as retrieving data.

The POST method, on the other hand, submits an entity (such as form data) to be processed by a specified resource. Unlike GET, a POST request is not idempotent, which means that making the same request multiple times may have different results. The POST method is often used to create, update or delete resources on the server.

## Q4

Templates in Flask are files containing HTML markup with placeholders for dynamic content. They allow developers to separate the presentation layer of their application from its logic.

The render_template() function is used in Flask to render a template and generate an HTML response that can be sent back to the client. It takes the name of the template file as its first argument, and any additional variables to be passed to the template as keyword arguments. The variables passed to the template can then be used to dynamically generate the HTML response.
